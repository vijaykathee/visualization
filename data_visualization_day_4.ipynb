```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as s
```

```python
df=pd.read_csv('diamonds.csv')
df.head()
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
d=df.select_dtypes(include='number').columns.tolist()
```

#### Write a Python program to display all Univariate Plots from the dataframe.


1.   Numeric based plots
2.   Categorical based plots




```python
#### code here
for col in d:
    s.boxplot(col,data=df)
    plt.show()
```

```python
for col in d:
    s.violinplot(col,data=df)
    plt.show()
```

```python
for col in d:
    s.histplot(df[col],kde=True)
    plt.show()
```

```python
#### code here
for col in df.select_dtypes(exclude='number').columns.tolist():
    s.countplot(col,data=df)
    plt.show()
```

###  Numeric to Categorical Plots

```python
s.violinplot(x='cut',y='price',data=df)
```

```python
s.barplot(x='cut',y='price',data=df)
```

### Categorical to Categorical Plots

```python
pd.crosstab(df['cut'],df['color']).plot(kind='bar')
plt.show()
```
