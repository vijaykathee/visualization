#### 2. What is box plot ?

-A boxplot is a standardized way of displaying the dataset based on the five-number summary: the minimum, the maximum, the sample median, and the first and third quartiles.
-A boxplot is use to identify the outliers in the dataset
-The area inside the box 50% of the data is known as the Inter Quartile Range.Calculated as – IQR = Q3-Q1
-Outliers are the data points below and above the lower and upper limit. The lower and upper limit is calculated as – Lower Limit = Q1 - 1.5*IQRUpper Limit = Q3 + 1.5*IQR
-The outliers which are outside this range of lower and upper limit.
